```python
# Imports
import os

# Data science and vis tools
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pingouin as pg

# Sklearn
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Custom packages
from sim_data import get_sim_data
```

```python
def pcorr(X_corr):
        X_inv = np.linalg.inv(X_corr)
        X_pcorr = -X_inv / np.outer(np.sqrt(np.diag(X_inv)), np.sqrt(np.diag(X_inv)))
        
        return X_pcorr
```

```python
#Import Simulated Data (Will take a while to run)
data = get_sim_data()
```

```python
#data partial
for i in range(len(data)):
    data[i]['pcorr'] = data[i]['corr'].apply(lambda x : pcorr(x))
```

```python
#data partial and fischer
for i in range(len(data)):
    data[i]['fpcorr'] = data[i]['corr'].apply(lambda x : np.arctanh(pcorr(x)))
```

```python
def kkn10(data):
    X = data['corr'].apply(lambda x : x.flatten()).to_list()
    y = data['gender'].to_list()
    
    # Generate Train Test Split
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0, test_size = 0.1)
    
    # Knn classifier
    knn10 = KNeighborsClassifier(n_neighbors = 10)
    
    knn10.fit(X_train, y_train)
    y_pred_2 = knn10.predict(X_test)
    
    #print("Accuracy with k=10", accuracy_score(y_test, y_pred_2)*100)
    return accuracy_score(y_test, y_pred_2)*100
```

```python
for i in range(len(data)):
    print("Accuracy with k=10 and separation of " + str((i + 1)/10) + ': ' + str(round(kkn10(data[i]), 1)) + '%')
```
